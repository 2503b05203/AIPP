Task-16.1

```python
import sqlite3

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create Books table
cursor.execute('''
CREATE TABLE Books (
    BookID INTEGER PRIMARY KEY AUTOINCREMENT,
    Title TEXT NOT NULL,
    Author TEXT NOT NULL,
    ISBN TEXT UNIQUE NOT NULL,
    PublicationYear INTEGER,
    Status TEXT NOT NULL DEFAULT 'Available'
);
''')

# Create Members table
cursor.execute('''
CREATE TABLE Members (
    MemberID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL,
    Address TEXT,
    PhoneNumber TEXT,
    Email TEXT UNIQUE
);
''')

# Create Loans table
cursor.execute('''
CREATE TABLE Loans (
    LoanID INTEGER PRIMARY KEY AUTOINCREMENT,
    BookID INTEGER NOT NULL,
    MemberID INTEGER NOT NULL,
    LoanDate TEXT NOT NULL,
    ReturnDate TEXT,
    FOREIGN KEY (BookID) REFERENCES Books(BookID),
    FOREIGN KEY (MemberID) REFERENCES Members(MemberID)
);
''')

# Commit the changes
conn.commit()

print("Database schema created successfully with Books, Members, and Loans tables.")


# --- Code to display schema ---
# Function to display table schema using PRAGMA table_info
def display_table_schema(table_name):
    print(f"\n--- Schema for {table_name} table ---")
    cursor.execute(f"PRAGMA table_info({table_name});")
    schema_info = cursor.fetchall()
    if not schema_info:
        print(f"No schema information found for {table_name}.")
        return

    # Print header
    print(f"{'CID':<4} {'Name':<15} {'Type':<10} {'NotNull':<8} {'PK':<3} {'Dflt_Value':<15}")
    print(f"{'----':<4} {'---------------':<15} {'----------':<10} {'--------':<8} {'---':<3} {'---------------':<15}")

    # Print rows
    for col in schema_info:
        # col format: (cid, name, type, notnull, dflt_value, pk)
        print(f"{col[0]:<4} {col[1]:<15} {col[2]:<10} {bool(col[3]):<8} {bool(col[5]):<3} {str(col[4]):<15}")

# Display schema for each table
display_table_schema('Books')
display_table_schema('Members')
display_table_schema('Loans')
```

```python
print("\n--- Loan Details ---")

# Construct the SQL query with JOINs
query = """
SELECT
    B.Title,
    M.Name AS MemberName,
    L.LoanDate,
    L.ReturnDate
FROM
    Loans AS L
JOIN
    Books AS B ON L.BookID = B.BookID
JOIN
    Members AS M ON L.MemberID = M.MemberID
"""

# Execute the query
cursor.execute(query)

# Fetch all results
results = cursor.fetchall()

# Get column names for header
column_names = [description[0] for description in cursor.description]

# Display results in a readable table format
if not results:
    print("No loan data found.")
else:
    # Print header
    print(" | ".join(column_names))
    print("-" * (sum(len(col) for col in column_names) + 3 * (len(column_names) - 1)))

    # Print rows
    for row in results:
        print(" | ".join(map(str, row)))
```

```python
import sqlite3

# Re-connect to the in-memory SQLite database and get cursor
# Note: This will recreate the in-memory database and lose previous data if the original 'conn' object was lost.
# If you intend to preserve data across cells, ensure 'conn' and 'cursor' are global or passed around.
# For this task, we assume the previous cells defining the schema and inserting data have established 'conn' and 'cursor' globally,
# but for robustness against kernel restarts, we re-establish them if they are not defined.

# Check if conn and cursor are already defined, if not, re-establish connection
if 'conn' not in globals() or 'cursor' not in globals():
    print("Re-establishing in-memory database connection and cursor...")
    conn = sqlite3.connect(':memory:')
    cursor = conn.cursor()

    # Re-create Books table
    cursor.execute('''
    CREATE TABLE Books (
        BookID INTEGER PRIMARY KEY AUTOINCREMENT,
        Title TEXT NOT NULL,
        Author TEXT NOT NULL,
        ISBN TEXT UNIQUE NOT NULL,
        PublicationYear INTEGER,
        Status TEXT NOT NULL DEFAULT 'Available'
    );
    ''')

    # Re-create Members table
    cursor.execute('''
    CREATE TABLE Members (
        MemberID INTEGER PRIMARY KEY AUTOINCREMENT,
        Name TEXT NOT NULL,
        Address TEXT,
        PhoneNumber TEXT,
        Email TEXT UNIQUE
    );
    ''')

    # Re-create Loans table
    cursor.execute('''
    CREATE TABLE Loans (
        LoanID INTEGER PRIMARY KEY AUTOINCREMENT,
        BookID INTEGER NOT NULL,
        MemberID INTEGER NOT NULL,
        LoanDate TEXT NOT NULL,
        ReturnDate TEXT,
        FOREIGN KEY (BookID) REFERENCES Books(BookID),
        FOREIGN KEY (MemberID) REFERENCES Members(MemberID)
    );
    ''')
    conn.commit()
    print("Database schema re-created.")

    # Re-insert sample data if database was re-created
    print("Re-inserting sample data...")
    books_data = [
        ('The Great Gatsby', 'F. Scott Fitzgerald', '978-0743273565', 1925, 'Available'),
        ('1984', 'George Orwell', '978-0451524935', 1949, 'Available'),
        ('To Kill a a Mockingbird', 'Harper Lee', '978-0061120084', 1960, 'Available')
    ]
    cursor.executemany("INSERT INTO Books (Title, Author, ISBN, PublicationYear, Status) VALUES (?, ?, ?, ?, ?)", books_data)

    members_data = [
        ('Alice Smith', '123 Main St', '555-1111', 'alice@example.com'),
        ('Bob Johnson', '456 Oak Ave', '555-2222', 'bob@example.com'),
        ('Charlie Brown', '789 Pine Ln', '555-3333', 'charlie@example.com')
    ]
    cursor.executemany("INSERT INTO Members (Name, Address, PhoneNumber, Email) VALUES (?, ?, ?, ?)", members_data)

    loans_data = [
        (1, 1, '2023-01-01', None), # Alice borrowed The Great Gatsby
        (2, 2, '2023-01-05', None), # Bob borrowed 1984
        (3, 3, '2023-01-10', '2023-01-25') # Charlie borrowed To Kill a Mockingbird and returned it
    ]
    cursor.executemany("INSERT INTO Loans (BookID, MemberID, LoanDate, ReturnDate) VALUES (?, ?, ?, ?)", loans_data)
    conn.commit()
    print("Sample data re-inserted.")


# Ensure foreign key constraints are enabled (if not already from previous steps)
cursor.execute("PRAGMA foreign_keys = ON;")
conn.commit()

print("\n--- Updating Book Status to 'FALSE' for Borrowed Books ---")

# SQL to update Status to 'FALSE' for books currently on loan (ReturnDate IS NULL)
# Note: Assuming 'FALSE' as a string is the desired status based on the prompt.
update_books_status_sql = """
UPDATE Books
SET Status = 'FALSE'
WHERE BookID IN (SELECT BookID FROM Loans WHERE ReturnDate IS NULL);
"""
cursor.execute(update_books_status_sql);
conn.commit()
print("Book statuses updated.")

# Display updated Books table to verify status changes
display_table_data('Books');

print("\n--- Safely Deleting a Member Record (Member ID 1 - Alice Smith) ---")

# We'll choose MemberID 1 (Alice Smith) to demonstrate safe deletion.
# Alice has an outstanding loan (BookID 1, The Great Gatsby).
member_to_delete_id = 1;

print(f"Attempting to safely delete Member with ID {member_to_delete_id}...");

# Step 1: Delete all loan records associated with the member
# This ensures referential integrity since the Loans table depends on Members.
print(f"First, deleting all loan records for Member ID {member_to_delete_id}...");
cursor.execute(f"DELETE FROM Loans WHERE MemberID = {member_to_delete_id};");
conn.commit()
print(f"Loan records for Member ID {member_to_delete_id} deleted.");

# Step 2: Now that dependent loan records are gone, delete the member record
print(f"Now, deleting Member with ID {member_to_delete_id}...");
cursor.execute(f"DELETE FROM Members WHERE MemberID = {member_to_delete_id};");
conn.commit()
print(f"Member with ID {member_to_delete_id} deleted successfully.");

# Display updated Members and Books tables to show the results
display_table_data('Members');
display_table_data('Loans'); # Display loans to show records were removed
display_table_data('Books'); # Display books again to show consistent state
```
